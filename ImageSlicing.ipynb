```python
%matplotlib inline
from matplotlib import pyplot as plt
import cv2
import numpy as np
import image_slicer
from PIL import ImageDraw, ImageFont,Image
```

```python
tiles = image_slicer.slice('timage.tiff',14,save='False')
```

```python
l = []
for tile in tiles:
    #t = Image.open(tile)
    overlay = ImageDraw.Draw(tile.image)
    l.append(tile)
    #print(overlay)
    overlay.text((5, 5), str(tile.number), (255, 255, 255),
    ImageFont.load_default())
image = image_slicer.join(tiles)
image.show()

#image_slicer.save_tiles(tiles)
```

```python
def show(im):
    img = cv2.imread(p)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
def findCoord(img):
    image  = cv2.imread(img)
    lgreen = np.array([0,0,0])
    ugreen = np.array([255,255,255])
    mask   = cv2.inRange(image,lgreen,ugreen)
    coord = cv2.findNonZero(mask)
    print(img, coord)
```
